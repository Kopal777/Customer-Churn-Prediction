```python
import pandas as pd
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.metrics import recall_score
from imblearn.combine import SMOTEENN
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import roc_auc_score, roc_curve
import matplotlib.pyplot as plt
from sklearn.svm import SVC
```

```python
df = pd.read_csv("tel_churn.csv")
df.head()
```

```python
df = df.drop('Unnamed: 0', axis = 1)
df.head()
```

```python
# Creating X and Y variables(as in features and prediction)

x = df.drop('Churn', axis = 1)
x
```

```python
y = df['Churn']
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

### Decision Tree Classifier without oversampling

```python
model_dt = DecisionTreeClassifier(criterion='gini', random_state=10, max_depth=6, min_samples_leaf=8)
model_dt.fit(x_train, y_train)
y_pred = model_dt.predict(x_test)
y_pred
```

```python
print(confusion_matrix(y_test, y_pred))
```

```python
print(classification_report(y_test, y_pred, labels=[0,1]))
```

We can observe that the accuracy was found to be 78% due to imbalanced dataset.

```python
# Probabilities for class 1
y_pred_proba_wdt = model_dt.predict_proba(x_test)[:, 1]  
# 2. Calculate the ROC-AUC score
roc_auc = roc_auc_score(y_test, y_pred_proba_wdt)
print(f"ROC-AUC Score Decision Trees without Oversampling: {roc_auc: .4f}")
```

### Decision Tree with oversamplig using SMOTEENN
We make use of SMOTTEENN 
(SMOTE:Synthetic Minority Over-sampling TEchnique &
EEN:Editorial Nearest Neighbors) 
used to address a class imbalance in a dataset by oversampling the data.
Over-sampling using SMOTE and cleaning using ENN.

```python
sme = SMOTEENN()
x_resampled, y_resampled = sme.fit_resample(x, y)
xr_train, xr_test, yr_train, yr_test = train_test_split(x_resampled, y_resampled, test_size = 0.2, random_state=42)
model_dt_smote = DecisionTreeClassifier(criterion='gini', random_state=10, max_depth=6, min_samples_leaf=8)
model_dt.fit(xr_train, yr_train)
```

```python
yr_pred = model_dt.predict(xr_test)
yr_pred
```

```python
print(classification_report(yr_test, yr_pred, labels=[0,1]))
```

```python
# Probabilities for class 1
y_pred_proba_dt = model_dt.predict_proba(xr_test)[:, 1]  
# 2. Calculate the ROC-AUC score
roc_auc = roc_auc_score(yr_test, y_pred_proba_dt)
print(f"ROC-AUC Score Decision Trees with Oversampling: {roc_auc: .4f}")
```

### Random Forest Classifier with SMOTEEN and Hyperparameter Tuning

Accuracy - Classification Report and Confusion Matrix

Hyperparameter tuning - GridSearchCV

```python
sme = SMOTEENN()
x_resampled, y_resampled = sme.fit_resample(x, y)
xr_train, xr_test, yr_train, yr_test = train_test_split(x_resampled, y_resampled, test_size = 0.2, random_state=42)
```

```python
param_grid = {
    'n_estimators': [50, 100, 200],    # Number of trees
    'max_depth': [4, 6, 8, 10],        # Maximum depth of the trees
    'min_samples_split': [2, 5, 10],   # Minimum number of samples required to split a node
    'min_samples_leaf': [1, 2, 4, 8],  # Minimum number of samples required at a leaf node
    'criterion': ['gini', 'entropy']   # Splitting criterion
}

rf_clf = RandomForestClassifier(random_state=42)
```

```python
grid_search = GridSearchCV(estimator=rf_clf, param_grid=param_grid, cv=5, n_jobs=-1, verbose=2)
```

```python
grid_search.fit(xr_train, yr_train)
```

```python
# rf_clf.fit(xr_train_rf, yr_train_rf)
best_rf_clf = grid_search.best_estimator_
print("Best Parameters: ", grid_search.best_params_)
```

```python
y_pred_rf = best_rf_clf.predict(xr_test)
y_pred_rf
```

```python
print(classification_report(yr_test, y_pred_rf, labels=[0,1]))
```

```python
print(confusion_matrix(yr_test, y_pred_rf))
```

Random Forest Classifier gives a much better result as compared to the other algorithms in terms of precision, recall, f1-score and accuracy.

```python
# Probabilities for class 1
y_pred_proba_rf = best_rf_clf.predict_proba(xr_test)[:, 1]  
# 2. Calculate the ROC-AUC score
roc_auc = roc_auc_score(yr_test, y_pred_proba_rf)
print(f"ROC_AUC Score Random Forest with Oversampling: {roc_auc: .4f}")
```

##### Random Forest Classifier - Accuracy calculated using Cross-Validation

```python
from sklearn.model_selection import cross_val_score

rf_clf = RandomForestClassifier(criterion= 'entropy', max_depth= 10, min_samples_leaf= 1, min_samples_split= 2, n_estimators= 100)
```

```python
cv_scores = cross_val_score(rf_clf, x_resampled, y_resampled, cv=10, scoring='accuracy')
print(f"Cross-Validation Accuracy: {cv_scores.mean():.2f}")

cv_f1_scores = cross_val_score(rf_clf, x_resampled, y_resampled, cv=10, scoring='f1')
print(f"Cross-Validation F1 Score: {cv_f1_scores.mean():.2f}")
```
